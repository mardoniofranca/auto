```python
import multiprocessing

from gensim.models import Word2Vec
```

```python
cores = multiprocessing.cpu_count() # Count the number of cores in a computer
print(cores)
```

```python
len(l_sentence)
```

```python
count = 0
for l in l_sentence:
    count = count + len(l)
print(count)
```

```python
w2v_model = Word2Vec(min_count=20,
                     window=2,
                     size = 300,
                     sample=6e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007, 
                     negative=20,
                     workers=cores-1)
```

```python
w2v_model.build_vocab(l_sentence, progress_per=10000)
```

```python
w2v_model.train(l_sentence, total_examples=len(l_sentence), epochs=30, report_delay=1)
```

```python
w2v_model.init_sims(replace=True)
```

```python
w2v_model.wv.most_similar(positive=["joão"])
```

```python
w2v_model.wv.most_similar(positive=["chicó"])
```

```python
w2v_model.wv.most_similar(positive=["joãogrilo"])
```

```python
def tsne_plot(model):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
tsne_plot(w2v_model)
```

```python
from gensim.models import word2vec

corpus = l_sentence


model = word2vec.Word2Vec(corpus, size=100, window=20, min_count=200, workers=4)
```

```python
tsne_plot(model)
```

```python
# A more selective model
model = word2vec.Word2Vec(corpus, size=100, window=20, min_count=500, workers=4)
tsne_plot(model)
```

```python
# A less selective model
model = word2vec.Word2Vec(corpus, size=100, window=20, min_count=100, workers=4)
tsne_plot(model)
```
